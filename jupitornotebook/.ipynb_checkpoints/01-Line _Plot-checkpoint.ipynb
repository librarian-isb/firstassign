# Adding Limits

```python
# Add libraies
import seaborn as sns
import matplotlib.pyplot as plt

# load datasets

phool = sns.load_dataset("iris")
phool
# Draw a line Plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolo ka Plot")
# set lim of graph
plt.xlim(2)
plt.ylim(1)
plt.show()
```

# Set Style

```python
# Add libraies
import seaborn as sns
import matplotlib.pyplot as plt
set_style (style="None", rc="None")
# load datasets

phool = sns.load_dataset("iris")
phool
# Draw a line Plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolo ka Plot")
# set style of graph
sns.set_style("dark")
plt.show()
```

# size of Figure

```python
# Add libraies
import seaborn as sns
import matplotlib.pyplot as plt

# load datasets

phool = sns.load_dataset("iris")
# change Figure size
plt.figure(figsize=(10,8))
# Draw a line Plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolo ka Plot")
# set style of graph
sns.set_style("dark")
plt.show()
```
