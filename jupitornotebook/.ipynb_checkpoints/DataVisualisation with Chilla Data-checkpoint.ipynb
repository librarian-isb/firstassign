```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#import data from file
chilla = pd.read_csv("Chilla_data2.csv")
# change Figure size
plt.figure(figsize=(10,8))
sns.lineplot(x="Gender", y="Your Weight in kg? (float)", data=chilla)
plt.title("Chilla kay Participants")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#import data from file
chilla = pd.read_csv("Chilla_data2.csv")
# change Figure size
plt.figure(figsize=(10,8))
sns.barplot(x="Gender", y="Your Weight in kg? (float)", data=chilla)
plt.title("Gender and Weight Plot")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy

#import data from file
chilla = pd.read_csv("Chilla_data2.csv")
# change Figure size
plt.figure(figsize=(8,6))
sns.barplot(x="Research/Working experience (Float/Int) years", y="Gender", hue="Your Weight in kg? (float)",data=chilla)
plt.title("Gender and Weight Plot")
plt.show()
```
