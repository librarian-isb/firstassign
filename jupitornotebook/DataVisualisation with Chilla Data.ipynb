```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#import data from file
chilla = pd.read_csv("Chilla_data2.csv")
sns.barplot(x="Gender", y="age", hue="area", data=chilla, estimator=np.mean)
chilla
plt.title("Gender and Area Plot")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#import data from file
chilla = pd.read_csv("Chilla_data2.csv")
sns.barplot(x="Gender", y="age", data=chilla,
            linewidth=3, facecolor(1,1,1,0),
           errcolor="1", edgecolor="1")
plt.title("Gender and Area Plot")
plt.show()
```
